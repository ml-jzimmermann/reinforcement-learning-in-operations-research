```python
from src.model import EmbeddingDQN
from src.agent import QAgent
import gym
import torch.nn.functional as F
import torch.optim as optimizer
import torch
```

```python
hyperparameters = {
    # training
    'batch_size': 256,
    'learning_rate': 0.001,
    'lr_decay': 5000,
    'lr_end': 0.0001,
    'optimizer': optimizer.Adam,
    'loss': F.smooth_l1_loss,
    # reinforcement & environment
    'eps_end': 0.05,
    'eps_start': 1,
    'eps_decay': 500,
    'gamma': 0.95,
    'target_update': 20,
    'num_episodes': 10001,
    'memory_capacity': 50000,
    'warmup_episodes': 10,
    'save_freq': 2000,
    'max_steps_per_episode': 100,
    # pytorch
    'np_seed': 100,
    'device': 'cuda',
    'save_model': True,
    'dtype': torch.long,
    'ylim': (-300, 120),
}
```

```python
taxi = gym.make('Taxi-v3').env
taxi.seed(100)
```

```python
agent = QAgent(env=taxi, model=EmbeddingDQN, config=hyperparameters)
agent.compile()
agent.fit()
```

```python
agent.play(verbose=False, sleep=0.3, max_steps=25)
```
